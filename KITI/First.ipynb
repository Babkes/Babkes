```python
import pandas as pd
from urllib.parse import urlsplit
```

```python
#Reading

df_gr=pd.read_excel('GR.xlsx')
df_jl=pd.read_excel('JL.xlsx')
df_lp=pd.read_excel('LP.xlsx')

df_gr=df_gr.fillna(0)
df_jl=df_jl.fillna(0)
df_lp=df_lp.fillna(0)
```

```python
#converting to dicts

l_gr=df_gr.to_dict(orient='records')
l_jl=df_jl.to_dict(orient='records')
l_lp=df_lp.to_dict(orient='records')
```

```python
#Selecting Innovation rows

i_lp=[]
for item in l_lp:
    if item['INNO']==1:
        i_lp.append(item)
        
i_gr=[]
for item in l_gr:
    if item['Intangibles_PRE_market_stage\n']==1 or item['Intangibles_impossible_to_say']==1 or item['Intangibles_market_stage']==1 or item['Tangibles_PRE_market_stage']==1 or item['Tangibles_impossible_to_say']==1 or item['Tangibles_market_stage']==1:
        i_gr.append(item)
i_jl=[]
for item in l_jl:
    if item['Intangibles_PRE_market_stage\n']==1 or item['Intangibles_impossible_to_say']==1 or item['Intangibles_market_stage']==1 or item['Tangibles_PRE_market_stage']==1 or item['Tangibles_impossible_to_say']==1 or item['Tangibles_market_stage']==1:
        i_jl.append(item)    
```

```python
def overlap(list1, list2, list3):
    #123
    lall=[]
    for i in list1:
        if i in list2 and i in list3:
            lall.append(i)
    
    for i in list2:
        if i in list1 and i in list3:
            lall.append(i)
        
    for i in list3:
        if i in list1 and i in list2:
            lall.append(i) 
    lall=list(set(lall))
    
    print('Overlap 1-2-3: ', len(lall))
    print(lall)
    
    
    
    #12
    fs=[]
    for i in list1:
        if i in list2:
            fs.append(i)
    for i in list2:
        if i in list1:
            fs.append(i)          
    
    fs=list(set(fs))
    print('Overlap 1-2: ', len(fs))
    print(fs)
    
    #13
    ft=[]
    for i in list1:
        if i in list3:
            ft.append(i)
    for i in list3:
        if i in list1:
            ft.append(i)
    
    ft=list(set(ft))
    print('Overlap 1-3: ', len(ft))
    print(ft)
    
    #23
    st=[]
    for i in list2:
        if i in list3:
            st.append(i)
    for i in list3:
        if i in list2:
            st.append(i)
    st=list(set(st))
    print('Overlap 2-3: ', len(st))
    print(st)
```

```python
#Individual URLs
url_gr=[]
for i in i_gr:
    url_gr.append(i['site_url'])
    
url_jl=[]
for i in i_jl:
    url_jl.append(i['site_url'])

alphabet='abcdefghjklmnopqrstuvwyxzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890'
alfa2=[]
for i in alphabet:
    alfa2.append(i)
alfa2=tuple(alfa2)

durl_gr=[]
for url in url_gr:
    parts=urlsplit(url)
    d={'domain': '', 'site_path': ''}
    for part in parts:
        if part.startswith('http'):
            a=2
        elif part.startswith(alfa2) and '.' in part:
            d1={'domain': part}
            d.update(d1)
        elif part.startswith('/'):
            d2={'site_path': part}
            d.update(d2)
        else:
            d3={'oth': part}
            d.update(d3)
    durl_gr.append(d)
    
durl_jl=[]
for url in url_jl:
    parts=urlsplit(url)
    d={'domain': '', 'site_path': ''}
    for part in parts:
        if part.startswith('http'):
            a=2
        elif part.startswith(alfa2) and '.' in part:
            d1={'domain': part}
            d.update(d1)
        elif part.startswith('www'):
            d11={'domain': part}
            d.update(d11)
        elif part.startswith('/'):
            d2={'site_path': part}
            d.update(d2)
        else:
            d3={'oth': part}
            d.update(d3)
    durl_jl.append(d)

```

```python
d_gr={}
for item in durl_gr:
    if item['domain'] not in d_gr.keys():
        d={item['domain']: [item['site_path']]}
        d_gr.update(d)
    else:
        d_gr[item['domain']].append(item['site_path'])
d_jl={}
for item in durl_jl:
    if item['domain'] not in d_jl.keys():
        d={item['domain']: [item['site_path']]}
        d_jl.update(d)
    else:
        d_jl[item['domain']].append(item['site_path'])
```

```python
comms=[]

for key in d_gr.keys():
    if key in d_jl.keys():
        gr=d_gr[key]
        jl=d_jl[key]
        
        same=[]
        for item in gr:
            if item in jl:
                same.append(item)
        diffy=[]
        for item in gr:
            if item not in jl:
                dx={'only GR': item}
                diffy.append(dx)
        for item in jl:
            if item not in gr:
                dx={'only JL': item}
                diffy.append(dx)
                
        d={'domain': key, 'same': same, 'different': diffy}
        comms.append(d)
    else:
        d={'domain': {'only GR':key}, 'different': {'only GR': d_gr[key]}, 'same':''}
        
for key in d_jl.keys():
    if key not in d_gr.keys():
        d={'domain': {'only JL': key}, 'different':{'only JL': d_jl[key]}, 'same':''}
        comms.append(d)
        
```
